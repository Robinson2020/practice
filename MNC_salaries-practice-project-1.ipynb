```python
df1_salary.drop(["Name","Height",],axis=1,inplace=True)
df1_salary
```

```python
# check skewness
df1_salary.skew()
```

```python
# Treating using Log
import numpy as np
for col in df1_salary.columns:
    if df1_salary.skew().loc[col]>0.55:
        df1_salary[col]=np.log1p(df1_salary[col])
```

```python
#lets again check the skewness
df1_salary.skew()
```

```python
import seaborn as sns
sns.pairplot(df1_salary)
plt.show()
```

```python
#check for outliers
from scipy.stats import zscore
z_score=abs(zscore(df1_salary))
print(df1_salary.shape)
df1_salary_final=df1_salary.loc[(z_score<3).all(axis=1)]
print(df1_salary_final.shape)
df1_salary_final
```

```python
#separating the input variable
import numpy as np
df_x=df1_salary_final.drop(columns=["Salary"])
y=df1_salary_final['Salary']
```

```python
#sacaling down using standardscaler
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(df_x)
x=pd.DataFrame(x,columns=df_x.columns)
```

```python
x.skew()
x
```

```python
y.skew()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn import linear_model
max_r_score=-10
for r_state in range(42,101):
    x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=r_state,test_size=0.20)
    regr=linear_model.LinearRegression()
    regr.fit(x_train,y_train)
    y_pred=regr.predict(x_test)
    r2_scr=r2_score(y_test,y_pred)
    print("r2_score corresponding to random state: ",r_state,"is: ",r2_scr)
    if r2_scr>max_r_score:
        max_r_score=r2_scr
        final_r_state=r_state
    print()
    print()
    print("max r2 score corresponding to ", final_r_state,"is",max_r_score)
```

```python
#max r2 score corresponding to  48 is 0.17305850363834363
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=50,test_size=.25)
lr=linear_model.LinearRegression()
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
```

```python
# Lets evaluate RMSC and r2 score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
import numpy as np
print("r2 score is : ",r2_score(y_test,y_pred))
print("RMSC is : ",np.sqrt(mean_squared_error(y_test,y_pred)))
```
